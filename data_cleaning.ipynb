```python
def add_oceanic():
    for chr in np.arange(1, 23):
        file = pd.read_csv("ancestry_results/covid_ref_chr{}_rfmix.msp.tsv".format(chr), sep="\t", skiprows=1)
        new_file = file.copy()
        new_file['N319.0'] = (np.ones(len(file.index))*4).astype(int)
        new_file['N319.1'] = (np.ones(len(file.index))*4).astype(int)
        new_file.to_csv("ancestry_results_with_oceanic/covid_ref_chr{}_rfmix.msp.tsv".format(chr), index=False, sep="\t")
```

```python
def create_map_file(path1, path2):
    add_oceanic()
    map_file = pd.DataFrame(columns = ["cM", "chr"])
    for chr in np.arange(1, 23):
        file_path = path1+str(chr)+path2
        chr_file = pd.read_csv(file_path, sep="\t")
        sixes2twos_msp(chr, chr_file)
        chr_map = chr_file[["sgpos", "#chm"]].rename(columns={"sgpos":"cM", "#chm":"chr"})
        chr_map['chr'] = chr_map.apply(lambda x: x['chr'][3:], 1) 
        map_file = pd.concat([map_file, chr_map])
        print("chr {} map completed".format(chr))
    map_file.to_csv("map_file.tsv", index=False, sep="\t")

def sixes2twos_msp(chr, chr_file):
    cols = [col for col in chr_file.columns if col == "sgpos" or col.startswith("S") and not col.startswith("SS")
            or col.startswith("N") and not col.startswith("NA")]
    new_df = chr_file[cols]
    participant_cols = [col for col in cols if col != "sgpos"]
    new_cleaned_df = new_df.copy()
    sixes2twos = new_df.loc[:, participant_cols].mask(new_df[participant_cols] == 6, 2)
    new_cleaned_df.loc[:, participant_cols] = sixes2twos
    new_cleaned_df.to_csv("cleaned_ancestry_results/chr{}.msp.tsv".format(chr), index=False, sep="\t")
    print("chr {} cleaning completed".format(chr))
```

```python
### Converts .msp file to .vcf format
###
### msp_file: the original file ending in .msp
### output_file: file path and name for the resulting .vcf file
### ancestry: integer representing the ancestry of interest to binary encode the data
### EAS=1, EUR/WAS=2, NAT=3, OCE=4, SAS=5

def msp2vcf(msp_file, output_file, ancestry):
    msp = open(msp_file)
    chromosome = msp_file[-9]
    output = open(output_file, 'w')
    output.write("##fileformat=VCFv4.2 \n##filedate=20220513 \n##source={}\n".format(msp_file))
    output.write('##FORMAT=<ID=GT,Number=1,Type=String,Description="Genotype">\n')
    for line in msp:
        bits = line.split()
        # write the participant ID's to the file, joining their entries
        # (eg. S100.0, S100.1 -> S100)
        if bits[0] == "spos":
            line_str = "#CHROM\tPOS\tID\tREF\tALT\tQUAL\tFILTER\tINFO\tFORMAT"
            line_str = "\t".join([line_str, bits[1][:-2]])
            for i in np.arange(3, len(bits)):
                if (i%2 == 1):
                    line_str = "\t".join([line_str, bits[i][:-2]])
        # write the binary coded ancestries for each participant
        # for example, if ancestry=2, then 3 2 would be coded as 0|1 in the output file
        else:
            line_str = "\t".join([chromosome, bits[0], '.', '.', '.', '.', 'PASS', '.', 'GT'])
            for i in np.arange(1, len(bits)):
                binary = int(int(bits[i]) == ancestry)
                if (i%2 == 1):
                    line_str += "\t" + str(binary) + "|"
                else:
                    line_str += str(binary)
        output.write(line_str + "\n")
```

```python
### Converts all msp files to vcf format by calling msp2vcf()
###
### input_dir: string path to the directory with the .msp files
### output_dir: string path to the directory in which to store .vcp files
### ancestries: list of ancestry integers of interest (in our case, [1,2,3,4,5])

def convert_all2vcf(input_dir, output_dir, ancestries):
    for chrom in np.arange(1, 23):
        for ancestry in ancestries:
            msp_file = input_dir + "chr{}.msp.tsv".format(chrom)
            output_file = output_dir + "covid_ref_chr{}_ancestry{}_rfmix.vcf".format(chrom, ancestry)
            msp2vcf(msp_file, output_file, ancestry)
            print("Converted msp->vcf for chromosome #{}, ancestry #{}".format(chrom, ancestry))
```

```python
convert_all2vcf("cleaned_pre_vcf/", "vcf_files/", [1,2,3,4,5])
```
